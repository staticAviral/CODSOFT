importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading dataset to a pandas dataframe
credit_card_data = pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Import necessary libraries
from sklearn.impute import SimpleImputer

# Define columns with missing values
columns_with_missing = credit_card_data.columns[credit_card_data.isnull().any()]

# Check if there are missing values in these columns
if not columns_with_missing.empty:
    # Create a SimpleImputer with the 'median' strategy
    imputer = SimpleImputer(strategy='median')
    
    # Apply imputation to columns with missing values
    credit_card_data[columns_with_missing] = imputer.fit_transform(credit_card_data[columns_with_missing])
    
    # Verify that missing values have been imputed
    print(credit_card_data.isnull().sum())
else:
    print("No missing values found in the DataFrame.")
```

```python
#distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced
0 --> normal
1 --> fraudulent

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

##under-sampling

build a sample dataset containing similar distribution of normal transactions and fraudulent transactions

```python
legit_sample = legit.sample(n=106)
```

concatenating two data frames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
# splitting data into features and targets
X = new_dataset.drop(columns = 'Class', axis = 1) #--> features
Y = new_dataset['Class'] #--> targets
```

```python
# split the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
#Model Training
model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

```python
#Model Evaluation
#Accuracy on training data
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred, Y_train)
```

```python
print("Accuracy on training data: ", training_data_accuracy)
```

```python
#accuracy on test data
X_test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred, Y_test)
```

```python
print("Accuracy on training data: ", test_data_accuracy)
```
